### Stemming

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize

def implement_porterStemmer(data):
    stemmer = PorterStemmer()
    vocab = word_tokenize(data)
    for word in vocab:
        print(word,"---->",stemmer.stem(word))
        
def implement_lancaster(data):
     stemmer = LancasterStemmer()
     vocab = word_tokenize(data)
     for word in vocab:
        print(word,"---->",stemmer.stem(word))
        
def implement_snowball(data):
    stemmer = SnowballStemmer(language="english")
    vocab = word_tokenize(data)
    for word in vocab:
        print(word,"---->",stemmer.stem(word))
    
stemmers = {
    "porter": implement_porterStemmer,
    "lancaster": implement_lancaster,
    "snowball": implement_snowball
}
    
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

documents = [
    "running runs ran easily fair fairness caring cared",
    "studies studying studied study goes going gone went",
    "The striped bats are hanging on their feet and ate best fishes",
    "She was running faster than anyone and had better results in studies",
    "relational conditional rational happiness organization"
]

for document in documents:
    print(f"Sentence: {document}")
    print("-" * len(document))
    
    print("Lemmtisation Results (No POS)")
    print("-" * len("lemmtisation results (no pos)"))
    
    words = word_tokenize(document)
    
    lemmatized = [lemmatizer.lemmatize(word) for word in words]
    
    print("WordNet Lemmatizer")
    print(lemmatized)
    
    print("-" * len(document))
```
