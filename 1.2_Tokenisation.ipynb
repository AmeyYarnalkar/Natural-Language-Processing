```python
corpus = """
I love learning NLP. NLP is fun.
I love coding in Python.
Python is great for NLP and Machine Learning.
"""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

# from corpus to documents
documents = sent_tokenize(corpus)
print(documents)
```

```python
# # from corpus to words
vocab = word_tokenize(corpus)
print(vocab)
```

```python
# from documents to words
for document in documents:
    print(word_tokenize(document))
```

```python
from nltk.tokenize import wordpunct_tokenize, TreebankWordTokenizer
print(wordpunct_tokenize(corpus)) # every punctuation is separate even 'letter -> ' + letter
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus) # only last full stop is separate else every other full stop is not
```
