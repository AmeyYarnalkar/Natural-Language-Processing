```python
corpus = [
    "I love NLP",
    "NLP is powerful"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectoriser = TfidfVectorizer(max_features=100, ngram_range=(2,3))
```

```python
X = vectoriser.fit_transform(corpus)
print(vectoriser.vocabulary_)
```

```python
print(X.toarray())
```
