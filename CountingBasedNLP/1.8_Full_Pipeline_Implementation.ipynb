```python
corpus = """
Artificial Intelligence (AI) is transforming industries across the globe. 
Companies like Google, Microsoft, and OpenAI are investing heavily in AI research and development.
In 2023, OpenAI released advanced language models that significantly improved natural language understanding.

India is emerging as a global technology hub, with cities like Bangalore, Hyderabad, and Pune leading innovation.
Many startups in India are working on machine learning, fintech, and healthcare solutions.
The Indian government has also launched initiatives to promote digital transformation and skill development.

Elon Musk, the CEO of Tesla and SpaceX, has been a strong advocate for artificial intelligence, while also warning about its risks.
Tesla focuses on electric vehicles and autonomous driving technology, whereas SpaceX is working on space exploration and satellite systems.

Education is evolving rapidly with the integration of technology.
Universities such as Stanford University and MIT are offering specialized courses in data science and artificial intelligence.
Online platforms like Coursera and edX provide access to high-quality education for students worldwide.

The healthcare sector is also benefiting from AI.
Machine learning models are being used to detect diseases, analyze medical images, and assist doctors in diagnosis.
Hospitals in the United States and Europe are increasingly adopting AI-driven solutions to improve patient care.

Climate change remains one of the biggest challenges of our time.
Organizations like the United Nations are working with governments to reduce carbon emissions and promote sustainability.
Renewable energy sources such as solar and wind power are gaining popularity as alternatives to fossil fuels.

Sports also generate a lot of data.
Players like Virat Kohli and Lionel Messi have millions of fans worldwide.
Teams use data analytics to improve performance, strategy, and decision-making.

In conclusion, technology, education, healthcare, and sustainability are interconnected domains that will shape the future.
The ability to analyze data and extract meaningful insights will be a critical skill in the coming years.
"""
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk import pos_tag, ne_chunk

# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('averaged_perceptron_tagger')
# nltk.download('maxent_ne_chunker')
# nltk.download('words')
```

```python
stop_words = set(stopwords.words('english'))
def handle_stopwords(token):
    temp = [word for word in token if word.lower() not in stop_words]
    return temp

# function to extract the information from the named entity
def extract_info(tree):
    relations = []
    for subtree in tree:
        if hasattr(subtree, "label"):
            entity_name = " ".join([token for token,pos in subtree.leaves()])
            entity_label = subtree.label()
            relations.append((entity_name, entity_label))
    return relations
```

```python
# tokenize the words
documents = sent_tokenize(corpus)
vocabulary = [word_tokenize(doc) for doc in documents]

# remove the stopping word from the tokens
filtered_tokens = []

#remove the stopping words
for word_tokens in vocabulary:
    filtered_tokens.append(handle_stopwords(word_tokens))
    
# doing the pos tagging
tags = []

for word_tokens in filtered_tokens:
    tags.append(pos_tag(word_tokens))
    
    
info = []
#doing the named entity relation ship
for word_tags in tags:
    tree = ne_chunk(word_tags)
    info.append(extract_info(tree))
```

```python
import json
from pathlib import Path

data = {}
file_path = Path().cwd() / "1.8_data.json"

for index, document in enumerate(documents):
    temp = {}

    if info[index]:
        for key, value in info[index]:   # safer
            temp[key] = value

    document = document.replace("\n", "")
    data[document] = temp

# write to file
with open(file_path, "w") as f:
    json.dump(data, f, indent=4)

print("JSON file created successfully!")
```

```python
import spacy

# load model
nlp = spacy.load("en_core_web_sm")

corpus = """
Artificial Intelligence (AI) is transforming industries across the globe.
Companies like Google, Microsoft, and OpenAI are investing heavily in AI research and development.
Elon Musk founded SpaceX and Tesla in the United States.
"""

# process entire text
doc = nlp(corpus)

# sentence-wise processing
for sent in doc.sents:
    print("\nSentence:", sent.text)
    print("-" * 50)

    # Tokens
    tokens = [token.text for token in sent]
    print("Tokens:", tokens)

    # Stopword removal
    filtered = [token.text for token in sent if not token.is_stop]
    print("After Stopword Removal:", filtered)

    # Lemmatization
    lemmas = [token.lemma_ for token in sent]
    print("Lemmatized:", lemmas)

    # POS Tagging
    pos_tags = [(token.text, token.pos_) for token in sent]
    print("POS Tags:", pos_tags)

    # Named Entity Recognition
    entities = [(ent.text, ent.label_) for ent in sent.ents]
    print("Entities:", entities)
    
result = []

for sent in doc.sents:
    sentence_data = {
        "sentence": sent.text,
        "entities": [
            {"text": ent.text, "label": ent.label_}
            for ent in sent.ents
        ]
    }
    result.append(sentence_data)

print(result)
```
