```python
from nltk import pos_tag_sents
from nltk.tokenize import word_tokenize

documents = [
    "running runs ran easily fair fairness caring cared",
    "studies studying studied study goes going gone went",
    "The striped bats are hanging on their feet and ate best fishes",
    "She was running faster than anyone and had better results in studies",
    "relational conditional rational happiness organization"
]

tokenised_docs = [word_tokenize(doc) for doc in documents]

print(pos_tag_sents(sentences=tokenised_docs))
```


##  POS Tags 

---

###  1. Noun

**Tags:** NN, NNS, NNP, NNPS

* **NN** → Singular noun → *dog*
* **NNS** → Plural noun → *dogs*
* **NNP** → Proper noun (singular) → *India*
* **NNPS** → Proper noun (plural) → *Indians*

---

###  2. Verb

**Tags:** VB, VBD, VBG, VBN, VBP, VBZ

* **VB** → Base form → *run*
* **VBD** → Past tense → *ran*
* **VBG** → Gerund (-ing form) → *running*
* **VBN** → Past participle → *eaten*
* **VBP** → Present (non-3rd person) → *run*
* **VBZ** → 3rd person singular → *runs*

---

###  3. Adjective

**Tags:** JJ, JJR, JJS

* **JJ** → Adjective → *fast*
* **JJR** → Comparative → *faster*
* **JJS** → Superlative → *fastest*

---

###  4. Adverb

**Tags:** RB, RBR, RBS

* **RB** → Adverb → *quickly*
* **RBR** → Comparative → *faster*
* **RBS** → Superlative → *fastest*

---

###  5. Pronoun

**Tags:** PRP, PRP$

* **PRP** → Personal pronoun → *he, she, they*
* **PRP$** → Possessive pronoun → *his, her, your*

---

###  6. Determiner

**Tag:** DT

* Articles / determiners → *the, a, this, those*

---

###  7. Preposition / Subordinating Conjunction

**Tag:** IN

* *in, on, of, that, because*

---

###  8. Coordinating Conjunction

**Tag:** CC

* *and, but, or*

---

###  9. Modal

**Tag:** MD

* *can, should, must, will*

---

### 10. Cardinal Number

**Tag:** CD

* *one, 2, 30*

---

###  11. Particle

**Tag:** RP

* *up, off, out, about*

---

###  12. Interjection

**Tag:** UH

* *oh, wow, hey*

---
